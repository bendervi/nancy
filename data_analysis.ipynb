### Compute the total amount of money traded and the use it as weights

```python
def money_to_number(x):
    if x in ["-", None, ""]:
        return 0
    x = x.replace("$", "")
    if "M" in x:
        return float(x.replace("M", "")) * 1_000_000
    if "K" in x:
        return float(x.replace("K", "")) * 1_000
    return float(x)
    
df["amount_num"] = df["amount"].apply(money_to_number)"

df['weights'] = df['amount_num'] / df['amount_num'].sum()

df
```

```python
# ta bort aktier med "N/A" / "None" ticker
valid_tickers = df[~df['ticker'].isin(['N/A', 'None'])]['ticker'].unique().tolist()
df = df[df['ticker'].isin(valid_tickers)]

# YFINANCE data nerladdnign
start_download = pd.to_datetime(df["date"].min())
end_download = pd.to_datetime("2025-01-01")

# ladda ner data på aktier som har handlats
market_data = yf.download(valid_tickers, start=start_download, end=end_download)["Open"]

# fyll luckor under helgen med fredags pris
market_data = market_data.ffill()



@dataclass
class Position:
    shares: float
    entry_price: float
    direction: str
    entry_date: str




@dataclass
class Portfolio:
    positions: dict = field(default_factory=dict)
    history: list = field(default_factory=list)



    def add_position(self, ticker: str, price: float, direction: str, date: str):
        if pd.isna(price):
            print(f"Skipping {ticker} on {date} (No Price Data)")
            return

        notional = 100.0  # $100 per signal
        shares = notional / price

        if direction.upper() == "SELL":
            shares = -shares  # kort position

        pos = Position(shares, price, direction.upper(), date)

        if ticker not in self.positions:
            self.positions[ticker] = []
        self.positions[ticker].append(pos)



    def value_on_date(self, date_ts):
        total_value = 0.0

        if date_ts not in market_data.index:
            return 0

        for ticker, lots in self.positions.items():
            if ticker not in market_data.columns:
                continue

            price = market_data.loc[date_ts, ticker]

            if pd.isna(price):
                continue


            for lot in lots:
                total_value += lot.shares * price


        return total_value



    def record_daily(self, date_ts):
        value = self.value_on_date(date_ts)
        self.history.append({"date": date_ts, "value": value})



    def get_history_df(self):
        return pd.DataFrame(self.history)



# skapa en portfölj
portfolio = Portfolio()

# sortera signal i order av datum
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

start_date = df["date"].min()
end_date = df["date"].max()
simulation_dates = pd.date_range(start=start_date, end=end_date)

for current_date in simulation_dates:
    string_date = current_date.strftime("%Y-%m-%d")

    # 1. updatera portfölj värde M2M (mark-to-market)
    portfolio.record_daily(current_date)

    # 2. kolla efter ny signal idag
    todays_signals = df[df["date"] == current_date]

    if todays_signals.empty:
        continue

    for _, row in todays_signals.iterrows():
        ticker = row["ticker"]
        direction = row["action"]

        # kolla pris och lägg till position i portfölj
        if current_date in market_data.index and ticker in market_data.columns:
            price = market_data.loc[current_date, ticker]
            portfolio.add_position(ticker, price, direction, string_date)

```

```python
history_df = portfolio.get_history_df()
history_df.set_index('date', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(history_df.index, history_df["value"], label="Portfolio Value")
plt.title("Portfolio Value")
plt.xlabel('Date')
plt.ylabel("Value ($)")
plt.grid(True)
plt.legend()
plt.show()
```
